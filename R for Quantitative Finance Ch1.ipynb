### 시계열 데이터 다루기

```r
install.packages('zoo')
```

```r
library('zoo')
```

```r
file_loc='C:/Users/sds/Documents/aapl.csv'
aapl <- read.zoo(file_loc, sep=',', header=T, format='%Y-%m-%d')
```

```r
plot(aapl, main="APPLE Closing Prices on NASDAQ", ylab='Price in USD', xlab='Date')
```

```r
aapl[which.max(aapl)] ## 최고가 확인
```

```r
ret_simple <- diff(aapl) / lag(aapl, k=-1) * 100 ## 단순 수익률
ret_cont <- diff(log(aapl)) * 100 ## 복리 수익률
```

```r
summary(coredata(ret_simple))
```

```r
ret_simple[which.min(ret_simple)]
```

```r
hist(ret_simple, breaks=100, main="Histogram of Simple Returns", xlab='%')
```

```r
hist(ret_cont, breaks=100, main="Histogram of Continuous Returns", xlab='%')
```

```r
aapl_2013 <- window(aapl, start='2013-01-01', end='2013-12-31')
aapl_2013 [which.max(aapl_2013)]
```

```r
quantile(ret_simple, probs=0.01)
```
